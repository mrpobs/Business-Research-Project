```python
# Open new jupyter notebook and import pandas
import pandas as pd

# All of the following files were provided by https://www.economia.gob.mx/datamexico/

# Import all the CSV files: Net international trade wih each country of the EU + European Community
# List of countries and file names
countries = [
    'Austria', 'Belgium', 'Bulgary', 'Croatia', 'Cyprus', 'Czech_Republic',
    'Denmark', 'Estonia', 'European_Community', 'Finland', 'France', 'Germany',
    'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg',
    'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia',
    'Spain', 'Sweden'
]
file_names = [
    'Net-International-Trade-with-Austria-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Belgium-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Bulgaria-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Croatia-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Cyprus-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Czech-Republic-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Denmark-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Estonia-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-European-Community-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Finland-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-France-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Germany-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Greece-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Hungary-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Ireland-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Italy-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Latvia-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Lithuania-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Luxembourg-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Malta-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Netherlands-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Poland-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Portugal-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Romania-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Slovakia-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Slovenia-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Spain-according-BCMM--Monthly-Flow.csv',
    'Net-International-Trade-with-Sweden-according-BCMM--Monthly-Flow.csv'
]

# Dictionary to store processed DataFrames
processed_dfs = {}

# Loop through each country and file
for country, file_name in zip(countries, file_names):
    # Read the CSV file
    df = pd.read_csv(file_name)
    
    # Add a 'Country' column with the country name
    df['Country'] = country

     # Drop the 'Time', 'Flow ID' and 'Month ID' columns
    df.drop('Time', axis=1, inplace=True)
    df.drop('Flow ID', axis=1, inplace=True)
    df.drop('Month ID', axis=1, inplace=True)
    
        # Store the processed DataFrame in the dictionary
    processed_dfs[country] = df

    # Print the processed DataFrame (optional)
    print(f'Processed DataFrame for {country}:')
    print(df.head())  # Print the first few rows of the DataFrame
    print()  # Blank line
```

```python
# We have saved and can access each processed dataframe using the processed_dfs dictionary and the key (country)
# processed_dfs['Austria']
```

```python
# Empty DataFrames for 'Exports' and 'Imports'
exports_Mexico_EU = pd.DataFrame()
imports_Mexico_EU = pd.DataFrame()

# Iterate over the dictionary
for key, df in processed_dfs.items():
    # Filter rows where 'Flow' column is 'Exports' and concatenate to exports_Mexico_EU
    exports_Mexico_EU = pd.concat([exports_Mexico_EU, df[df['Flow'] == 'Exports']])
    # Filter rows where 'Flow' column is 'Imports' and concatenate to imports_Mexico_EU
    imports_Mexico_EU = pd.concat([imports_Mexico_EU, df[df['Flow'] == 'Imports']])

# Sort all the imports and exports by 'Month' and 'Country'
imports_MX_EU_by_country = imports_Mexico_EU.sort_values(by=['Month', 'Country'])
exports_MX_EU_by_country = exports_Mexico_EU.sort_values(by=['Month', 'Country'])

# Reset the index and drop the old index
imports_MX_EU_by_country.reset_index(drop=True, inplace=True)
exports_MX_EU_by_country.reset_index(drop=True, inplace=True)
```

```python
# Now lets do another two datasets with the total sum of the countries for each 'Month'
# Copy the Datasets with new names
imports_MX_EU = imports_MX_EU_by_country.copy()
exports_MX_EU = exports_MX_EU_by_country.copy()

# Set the 'Country' column as a different constant for each
imports_MX_EU['Country']='EU to Mexico'
exports_MX_EU['Country']='Mexico to EU'
# Rename the 'Country' column as Trade
imports_MX_EU = imports_MX_EU.rename(columns={'Country': 'Trade Direction'})
exports_MX_EU = exports_MX_EU.rename(columns={'Country': 'Trade Direction'})

# Group by 'Month' and sum the 'Trade Value' within each group
imports_MX_EU = imports_MX_EU.groupby('Month', as_index=False).agg({'Flow': 'first', 'Trade Value': 'sum', 'Trade Direction': 'first'})
exports_MX_EU = exports_MX_EU.groupby('Month', as_index=False).agg({'Flow': 'first', 'Trade Value': 'sum', 'Trade Direction': 'first'})

imports_MX_EU.head()
```
